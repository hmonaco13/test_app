#Example of APP#

User should choose a state, crop and parameter within the crop to plot it over time.

```python
import pandas as pd
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/hmonaco13/test_app/main/A839DBF9-E471-391A-894E-E24CE44009FB.csv'
df = pd.read_csv(url)
df1 = df[['State','Year','Commodity','Data Item','Value']]

state = "CALIFORNIA" #@param ['ARIZONA', 'ARKANSAS', 'CALIFORNIA', 'COLORADO', 'IDAHO','ILLINOIS', 'INDIANA', 'KANSAS', 'KENTUCKY', 'MARYLAND',       'MICHIGAN', 'MISSISSIPPI', 'MISSOURI', 'MONTANA', 'NEBRASKA',       'NORTH CAROLINA', 'NORTH DAKOTA', 'OHIO', 'OKLAHOMA', 'OREGON',       'OTHER STATES', 'SOUTH DAKOTA', 'TENNESSEE', 'TEXAS', 'VIRGINIA',       'WASHINGTON', 'WISCONSIN', 'ALABAMA', 'DELAWARE', 'FLORIDA',       'GEORGIA', 'IOWA', 'LOUISIANA', 'MINNESOTA', 'NEW JERSEY',       'NEW MEXICO', 'NEW YORK', 'PENNSYLVANIA', 'SOUTH CAROLINA', 'UTAH',       'WEST VIRGINIA', 'WYOMING', 'CONNECTICUT', 'MAINE',       'MASSACHUSETTS', 'NEVADA', 'NEW HAMPSHIRE', 'RHODE ISLAND','VERMONT', 'ALASKA', 'HAWAII']
crop ="WHEAT" #@param ['WHEAT', 'CORN', 'COTTON', 'SOYBEANS']
param = "WHEAT - YIELD, MEASURED IN BU / ACRE" #@param ['WHEAT - YIELD, MEASURED IN BU / ACRE','SOYBEANS, IRRIGATED - YIELD, MEASURED IN BU / ACRE', 'CORN, GRAIN - YIELD, MEASURED IN BU / ACRE','SOYBEANS, NON-IRRIGATED - YIELD, MEASURED IN BU / ACRE']

df1 = df1[(df1['State'] == state) & (df1['Commodity'] == crop) & (df1['Data Item'] == param)]
df1['Year'] = pd.to_numeric(df1['Year'] ,errors='coerce')
df1['Value'] = pd.to_numeric(df1['Value'] ,errors='coerce')

fig,ax =plt.subplots(figsize=(12,9))
df1.loc[:,(['Year','Value'])].set_index('Year').plot(ax=ax);
ax.set_ylabel(str(param))
ax.set_xlabel('Years')
plt.title(str(param)+' in '+str(state), fontsize=16);
```
